```python
def bootstrap_method(data):
    B = 1000
    vertex_count = 100
    xb = np.random.choice(data, (vertex_count, B), replace=True)
    apply_along_axis_mean = np.apply_along_axis(np.mean, 1, xb)
    mean_mean = np.mean(apply_along_axis_mean)
    se_mean = np.std(apply_along_axis_mean)
    apply_along_axis_median = np.apply_along_axis(np.median, 1, xb)
    mean_median = np.mean(apply_along_axis_median)
    se_median = np.std(apply_along_axis_median)
    return mean_mean, se_mean, mean_median, se_median
```

```python
for i in range(1,6):
    bootstrap_res = bootstrap_method(data['y'+str(i)])
    print("for y= {}".format(i))
    print("mean= {}".format(bootstrap_res[0]))
    print("error mean= {}".format(bootstrap_res[1]))
    print("median= {}".format(bootstrap_res[2]))
    print("error median= {}".format(bootstrap_res[3]))
    print("========================================================================")
```
